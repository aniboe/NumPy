##List vs Nunpy Array

```python
py_list = [1,2,3]
print("python muliplication:",py_list * 3)

np_array = np.array((1,2,3)) #element wise multiplication
print("python arry(numpy) muliplication:",np_array * 3)



import time
start = time.time()
py_list = [i*2 for i in range(100000)]
print("\nList operation time:",time.time()-start)

start = time.time()
np_array = np.arange(100000) * 2 # "arange" works similar to "range" in for loop
print("\nNumpy operation time:",time.time()-start)
```

##Creating array from scratch

```python
from typing import Sequence
zeros = np.zeros((3,4))
print("zero array:\n",zeros)

ones = np.ones((2,3)) # value inside is order of matrix
print("\nones array:\n",ones)

full = np.full((2,2),7) # array with 2x2 order with elemrnt 7
print("\nfull array:\n",full)


random = np.random.random((2,3))
print("\nransom array:\n",random)

Sequence = np.arange(0,10,2) # start, stop, step
print("\nSequence array:\n",Sequence)
```


##Vector, Matrix, Tenor

```python
vector = np.array([1,2,3])
print("vetctor:",vector)

matrix = np.array([[1,2,3],
                   [4,5,6]])
print("\nmatrix:",matrix)

tensor = np.array([[[1,2],[3,4]],
                   [[5,6],[6,8]]])
print("\ntensor:",tensor)
```

##Array properties

```python
arr = np.array([[1,2,3],
              [4,5,6]])

print("Shape",arr.shape)
print("Dimension",arr.ndim)
print("Size",arr.size)
print("DataType",arr.dtype)


arr2 = np.array([[1,2,3],
              [4,5,6.4]]) #float value here
print("\nShape",arr2.shape)
print("Dimension",arr2.ndim)
print("Size",arr2.size)
print("DataType",arr2.dtype)


arr3 = np.array([[1,2,3],
              [4,5,"6.4"]])
print("\nDataType",arr3.dtype)
```

##Array Reshaping

```python
arr = np.arange(12)
print("Original array",arr)

reshaped = arr.reshape((3,4))# converts arr into 3x4 matrix
print("\nReshaped array:",reshaped)

#rerns copy of an array
flatten = reshaped.flatten()
print("\nFlatten array:",flatten)

#returns view instead of copy(unlike flatten())
raveled = reshaped.ravel()
print("\nRaveled array:",raveled)


#Transpose
transpose = reshaped.T
print("\nTranspose array:",transpose)
```
