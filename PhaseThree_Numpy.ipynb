##Advanced operation eith Business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("=====zomato sales analysis==========")

print("\nsales data shape :",sales_data.shape)

print("\nsample data for first three :", sales_data[:3]) #same as "sales_data[0:3]"
print("\nsample data for first three :", sales_data[:, 1:]) # sales_data["rows" : "column", "start" : "stop"]
print("\nsample data for first three :", sales_data[:, 2:]) # sales_data["rows" : "column", "start" : "stop"]
```

```python
#Total sales per year
print(np.sum(sales_data, axis=0))

yearly_total = np.sum(sales_data[:, 1:], axis=0)
print("\nyearly total",yearly_total)


# Minimum sales per resurant
min_sales = np.min(sales_data[:, 1:], axis = 1)
print("\nmin sales",min_sales)


# Max sale per year
max_sale = np.max(sales_data[:,1:], axis = 0)
print("\nmax sale",max_sale)


# Avg sales per resturent
# avg_sale = np.average(sales_data[:,1:], axis = 1)
avg_sale = np.mean(sales_data[:,1:], axis = 1)
print("\navg sale",avg_sale)
```

```python
cumulativeSum = np.cumsum(sales_data[:,1:],axis = 1)
print("\ncumulative Sum: ",cumulativeSum)

plt.figure(figsize = (7, 5))
plt.plot(np.mean(cumulativeSum, axis = 0))
plt.title("Avg cumsum sales accross all resturant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

##Vectors

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

# vector addition
print("Vector  addition:", vector1 + vector2)

# vector subtraction
print("\nVector  subtraction:", vector1 - vector2)

# vector multiplication
print("\nVector  multiplication:", vector1 * vector2)

# vector division
print("\nVector  division:", vector1 / vector2)

# dot product
print("\nDot product:", np.dot(vector1, vector2))



#Angle between vectors
angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print("\nAngle:",angle)
```

```python
# Vectorised operation
resturant_types = np.array(["briani","chinese","pizza","burger","caffe"])

vectorised_upper = np.vectorize(str.upper)
print("vectorised upper",vectorised_upper(resturant_types))
```
